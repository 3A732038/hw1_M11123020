```python
import pandas as pd
set_1=pd.read_csv("set_1.csv",sep='\t',names=["loss","root_mean_squared_error","mean_absolute_error","mean_absolute_percentage_error","val_loss","val_root_mean_squared_error","val_mean_absolute_error","val_mean_absolute_percentage_error"])
set_2=pd.read_csv("set_2.csv",sep='\t',names=["loss","root_mean_squared_error","mean_absolute_error","mean_absolute_percentage_error","val_loss","val_root_mean_squared_error","val_mean_absolute_error","val_mean_absolute_percentage_error"])
set_3=pd.read_csv("set_3.csv",sep='\t',names=["loss","root_mean_squared_error","mean_absolute_error","mean_absolute_percentage_error","val_loss","val_root_mean_squared_error","val_mean_absolute_error","val_mean_absolute_percentage_error"])
set_4=pd.read_csv("set_4.csv",sep='\t',names=["loss","root_mean_squared_error","mean_absolute_error","mean_absolute_percentage_error","val_loss","val_root_mean_squared_error","val_mean_absolute_error","val_mean_absolute_percentage_error"])
```

```python
import matplotlib.pyplot as plt
from math import sqrt
def picture(name):    
    # 對訓練過程的mse繪圖
    plt.figure(figsize =(8,5))
    plt.plot(range(1,len(set_1[name])+1),set_1[name],label='set_1')
    plt.plot(range(1,len(set_2[name])+1),set_2[name],label='set_2')
    plt.plot(range(1,len(set_3[name])+1),set_3[name],label='set_3')
    plt.plot(range(1,len(set_4[name])+1),set_4[name],label='set_4')
    # plt.xticks(range(1,len(model_train.history[val_loss])+1))
    plt.ylabel(f"{name}")
    plt.xlabel("epoch")
    if 'val' in name:
        plt.title("val predict hours-per-week ")
    else:
       plt.title("training predict hours-per-week ")     
    plt.grid(True)#有網格
    plt.legend()#可以印出label
    plt.show()
```

```python
picture('val_mean_absolute_percentage_error')
```
